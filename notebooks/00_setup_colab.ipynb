## **ü§ñ Passo 3: Testando o LLM**

Agora vamos testar se tudo est√° funcionando. √â como dar uma volta de teste no carro antes de uma viagem longa!

```python
# ü§ñ TESTANDO O LLM
def get_llm_colab():
    """Retorna o melhor LLM dispon√≠vel no Colab"""
    
    # Tentativa 1: OpenAI
    try:
        from langchain_openai import ChatOpenAI
        api_key = os.getenv("OPENAI_API_KEY")
        if api_key:
            print("ü§ñ Usando OpenAI GPT-3.5-turbo")
            return ChatOpenAI(
                model="gpt-3.5-turbo",
                temperature=0.7,
                api_key=api_key
            )
    except Exception as e:
        print(f"‚ö†Ô∏è  OpenAI n√£o dispon√≠vel: {e}")
    
    # Tentativa 2: Hugging Face
    try:
        from langchain_community.llms import HuggingFaceHub
        token = os.getenv("HUGGINGFACEHUB_API_TOKEN")
        if token:
            print("ü§ó Usando Hugging Face (google/flan-t5-base)")
            return HuggingFaceHub(
                repo_id="google/flan-t5-base",
                model_kwargs={"temperature": 0.7, "max_length": 512}
            )
    except Exception as e:
        print(f"‚ö†Ô∏è  Hugging Face n√£o dispon√≠vel: {e}")
    
    # Fallback: Simula√ß√£o
    print("üîÑ Usando modo simula√ß√£o (sem API)")
    return None

# Testando o LLM
print("üß™ Testando configura√ß√£o do LLM...")
llm = get_llm_colab()

if llm:
    try:
        response = llm.invoke("Diga 'Ol√°, seguran√ßa em IA!' em portugu√™s")
        print(f"‚úÖ Teste bem-sucedido! Resposta: {response}")
    except Exception as e:
        print(f"‚ö†Ô∏è  Erro no teste: {e}")
else:
    print("‚úÖ Modo simula√ß√£o ativado - voc√™ pode continuar o curso normalmente!")
```

## **üîß Passo 4: Verificando Ferramentas de Seguran√ßa**

Vamos testar se todas as ferramentas de seguran√ßa est√£o funcionando:

```python
# üîß TESTANDO FERRAMENTAS DE SEGURAN√áA
import hashlib
import secrets
import logging
from datetime import datetime
import json
import re
from collections import Counter

print("üîí Testando ferramentas de seguran√ßa...")

# Teste 1: Hash de senha
senha_teste = "minhasenha123"
hash_senha = hashlib.sha256(senha_teste.encode()).hexdigest()
print(f"‚úÖ Hash de senha: {hash_senha[:20]}...")

# Teste 2: Gera√ß√£o de token seguro
token_seguro = secrets.token_urlsafe(32)
print(f"‚úÖ Token seguro gerado: {token_seguro[:20]}...")

# Teste 3: Logging
logging.basicConfig(level=logging.INFO)
logger = logging.getLogger(__name__)
logger.info("Sistema de logging funcionando!")
print("‚úÖ Sistema de logging ativo")

# Teste 4: Regex para valida√ß√£o
email_teste = "teste@exemplo.com"
padrao_email = r'^[\w\.-]+@[\w\.-]+\.\w+$'
if re.match(padrao_email, email_teste):
    print("‚úÖ Valida√ß√£o de email funcionando")

print("\nüéâ Todas as ferramentas de seguran√ßa est√£o funcionando!")
```

## **üìä Passo 5: Verificando Ferramentas de An√°lise**

Agora vamos testar as ferramentas de an√°lise e visualiza√ß√£o:

```python
# üìä TESTANDO FERRAMENTAS DE AN√ÅLISE
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
from sklearn.ensemble import IsolationForest

print("üìä Testando ferramentas de an√°lise...")

# Configurando estilo dos gr√°ficos
plt.style.use('default')
sns.set_palette("husl")

# Teste 1: Criando dados de exemplo
dados_teste = np.random.normal(0, 1, 1000)
df_teste = pd.DataFrame({
    'valor': dados_teste,
    'timestamp': pd.date_range('2024-01-01', periods=1000, freq='H')
})
print(f"‚úÖ DataFrame criado com {len(df_teste)} registros")

# Teste 2: Detec√ß√£o de anomalias
iso_forest = IsolationForest(contamination=0.1, random_state=42)
anomalias = iso_forest.fit_predict(dados_teste.reshape(-1, 1))
num_anomalias = (anomalias == -1).sum()
print(f"‚úÖ Detec√ß√£o de anomalias: {num_anomalias} anomalias encontradas")

# Teste 3: Gr√°fico simples
plt.figure(figsize=(10, 6))
plt.hist(dados_teste, bins=30, alpha=0.7, color='skyblue', edgecolor='black')
plt.title('Distribui√ß√£o de Dados de Teste')
plt.xlabel('Valor')
plt.ylabel('Frequ√™ncia')
plt.grid(True, alpha=0.3)
plt.show()

print("‚úÖ Ferramentas de an√°lise funcionando perfeitamente!")
```
