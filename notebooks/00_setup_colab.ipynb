## **ğŸ¤– Passo 3: Testando o LLM**

Agora vamos testar se tudo estÃ¡ funcionando. Ã‰ como dar uma volta de teste no carro antes de uma viagem longa!

```python
# ğŸ¤– TESTANDO O LLM
def get_llm_colab():
    """Retorna o melhor LLM disponÃ­vel no Colab"""
    
    # Tentativa 1: OpenAI
    try:
        from langchain_openai import ChatOpenAI
        api_key = os.getenv("OPENAI_API_KEY")
        if api_key:
            print("ğŸ¤– Usando OpenAI GPT-3.5-turbo")
            return ChatOpenAI(
                model="gpt-3.5-turbo",
                temperature=0.7,
                api_key=api_key
            )
    except Exception as e:
        print(f"âš ï¸  OpenAI nÃ£o disponÃ­vel: {e}")
    
    # Tentativa 2: Hugging Face
    try:
        from langchain_community.llms import HuggingFaceHub
        token = os.getenv("HUGGINGFACEHUB_API_TOKEN")
        if token:
            print("ğŸ¤— Usando Hugging Face (google/flan-t5-base)")
            return HuggingFaceHub(
                repo_id="google/flan-t5-base",
                model_kwargs={"temperature": 0.7, "max_length": 512}
            )
    except Exception as e:
        print(f"âš ï¸  Hugging Face nÃ£o disponÃ­vel: {e}")
    
    # Fallback: SimulaÃ§Ã£o
    print("ğŸ”„ Usando modo simulaÃ§Ã£o (sem API)")
    return None

# Testando o LLM
print("ğŸ§ª Testando configuraÃ§Ã£o do LLM...")
llm = get_llm_colab()

if llm:
    try:
        response = llm.invoke("Diga 'OlÃ¡, seguranÃ§a em IA!' em portuguÃªs")
        print(f"âœ… Teste bem-sucedido! Resposta: {response}")
    except Exception as e:
        print(f"âš ï¸  Erro no teste: {e}")
else:
    print("âœ… Modo simulaÃ§Ã£o ativado - vocÃª pode continuar o curso normalmente!")
```

## **ğŸ”§ Passo 4: Verificando Ferramentas de SeguranÃ§a**

Vamos testar se todas as ferramentas de seguranÃ§a estÃ£o funcionando:

```python
# ğŸ”§ TESTANDO FERRAMENTAS DE SEGURANÃ‡A
import hashlib
import secrets
import logging
from datetime import datetime
import json
import re
from collections import Counter

print("ğŸ”’ Testando ferramentas de seguranÃ§a...")

# Teste 1: Hash de senha
senha_teste = "minhasenha123"
hash_senha = hashlib.sha256(senha_teste.encode()).hexdigest()
print(f"âœ… Hash de senha: {hash_senha[:20]}...")

# Teste 2: GeraÃ§Ã£o de token seguro
token_seguro = secrets.token_urlsafe(32)
print(f"âœ… Token seguro gerado: {token_seguro[:20]}...")

# Teste 3: Logging
logging.basicConfig(level=logging.INFO)
logger = logging.getLogger(__name__)
logger.info("Sistema de logging funcionando!")
print("âœ… Sistema de logging ativo")

# Teste 4: Regex para validaÃ§Ã£o
email_teste = "teste@exemplo.com"
padrao_email = r'^[\w\.-]+@[\w\.-]+\.\w+$'
if re.match(padrao_email, email_teste):
    print("âœ… ValidaÃ§Ã£o de email funcionando")

print("\nğŸ‰ Todas as ferramentas de seguranÃ§a estÃ£o funcionando!")
```

## **ğŸ“Š Passo 5: Verificando Ferramentas de AnÃ¡lise**

Agora vamos testar as ferramentas de anÃ¡lise e visualizaÃ§Ã£o:

```python
# ğŸ“Š TESTANDO FERRAMENTAS DE ANÃLISE
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
from sklearn.ensemble import IsolationForest

print("ğŸ“Š Testando ferramentas de anÃ¡lise...")

# Configurando estilo dos grÃ¡ficos
plt.style.use('default')
sns.set_palette("husl")

# Teste 1: Criando dados de exemplo
dados_teste = np.random.normal(0, 1, 1000)
df_teste = pd.DataFrame({
    'valor': dados_teste,
    'timestamp': pd.date_range('2024-01-01', periods=1000, freq='H')
})
print(f"âœ… DataFrame criado com {len(df_teste)} registros")

# Teste 2: DetecÃ§Ã£o de anomalias
iso_forest = IsolationForest(contamination=0.1, random_state=42)
anomalias = iso_forest.fit_predict(dados_teste.reshape(-1, 1))
num_anomalias = (anomalias == -1).sum()
print(f"âœ… DetecÃ§Ã£o de anomalias: {num_anomalias} anomalias encontradas")

# Teste 3: GrÃ¡fico simples
plt.figure(figsize=(10, 6))
plt.hist(dados_teste, bins=30, alpha=0.7, color='skyblue', edgecolor='black')
plt.title('DistribuiÃ§Ã£o de Dados de Teste')
plt.xlabel('Valor')
plt.ylabel('FrequÃªncia')
plt.grid(True, alpha=0.3)
plt.show()

print("âœ… Ferramentas de anÃ¡lise funcionando perfeitamente!")
```
