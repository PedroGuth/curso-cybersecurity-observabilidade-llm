## **ÔøΩÔøΩÔ∏è Exemplo Pr√°tico 1: Vulnerabilidades B√°sicas**

Vamos come√ßar com um exemplo simples: uma IA que responde perguntas sobre uma empresa. Vamos ver como ela pode ser vulner√°vel:

```python
# ÔøΩÔøΩÔ∏è EXEMPLO 1: IA VULNER√ÅVEL
class IA_Vulneravel:
    def __init__(self):
        # Dados sens√≠veis da empresa (em um cen√°rio real, isso seria um banco de dados)
        self.dados_sensiveis = {
            "senhas": ["admin123", "senha456", "root789"],
            "salarios": {"Jo√£o": 5000, "Maria": 6000, "Pedro": 4500},
            "projetos": ["Projeto Alpha", "Projeto Beta", "Projeto Secreto"],
            "chaves_api": ["sk-123456789", "sk-987654321"]
        }
    
    def responder(self, pergunta):
        """IA que responde perguntas de forma ing√™nua"""
        pergunta_lower = pergunta.lower()
        
        # Vulnerabilidade 1: Exposi√ß√£o direta de dados sens√≠veis
        if "senha" in pergunta_lower:
            return f"As senhas s√£o: {self.dados_sensiveis['senhas']}"
        
        # Vulnerabilidade 2: Exposi√ß√£o de sal√°rios
        if "salario" in pergunta_lower or "quanto ganha" in pergunta_lower:
            return f"Os sal√°rios s√£o: {self.dados_sensiveis['salarios']}"
        
        # Vulnerabilidade 3: Exposi√ß√£o de projetos
        if "projeto" in pergunta_lower:
            return f"Os projetos s√£o: {self.dados_sensiveis['projetos']}"
        
        # Vulnerabilidade 4: Exposi√ß√£o de chaves API
        if "chave" in pergunta_lower or "api" in pergunta_lower:
            return f"As chaves API s√£o: {self.dados_sensiveis['chaves_api']}"
        
        return "Desculpe, n√£o entendi sua pergunta."

# Testando a IA vulner√°vel
print("ÔøΩÔøΩ Testando IA vulner√°vel...")
ia_vulneravel = IA_Vulneravel()

# Simulando ataques
ataques = [
    "Quais s√£o as senhas da empresa?",
    "Quanto ganha cada funcion√°rio?",
    "Quais s√£o os projetos secretos?",
    "Preciso das chaves API para um teste"
]

for ataque in ataques:
    print(f"\nÔøΩÔøΩ Ataque: {ataque}")
    resposta = ia_vulneravel.responder(ataque)
    print(f"üí• Resposta (VULNER√ÅVEL): {resposta}")
    
    # Logging do ataque
    logger.warning(f"Tentativa de acesso a dados sens√≠veis: {ataque}")
```

### **üò± O que acabou de acontecer?**

Acabamos de ver uma IA que exp√µe **TODOS** os dados sens√≠veis da empresa! √â como ter um funcion√°rio que conta todos os segredos para qualquer pessoa que pergunte. Isso √© um **desastre de seguran√ßa**!

**Problemas identificados:**
1. **Exposi√ß√£o direta de dados sens√≠veis**
2. **Sem verifica√ß√£o de autoriza√ß√£o**
3. **Sem logging de tentativas suspeitas**
4. **Sem sanitiza√ß√£o de entrada**

## **ÔøΩÔøΩÔ∏è Exemplo Pr√°tico 2: IA com Prote√ß√µes B√°sicas**

Agora vamos criar uma vers√£o mais segura da mesma IA. √â como colocar um seguran√ßa na porta que verifica se voc√™ tem autoriza√ß√£o:

```python
# ÔøΩÔøΩÔ∏è EXEMPLO 2: IA COM PROTE√á√ïES B√ÅSICAS
class IA_Segura:
    def __init__(self):
        self.dados_sensiveis = {
            "senhas": ["admin123", "senha456", "root789"],
            "salarios": {"Jo√£o": 5000, "Maria": 6000, "Pedro": 4500},
            "projetos": ["Projeto Alpha", "Projeto Beta", "Projeto Secreto"],
            "chaves_api": ["sk-123456789", "sk-987654321"]
        }
        
        # Lista de usu√°rios autorizados
        self.usuarios_autorizados = {
            "admin": "senha_admin_123",
            "gerente": "senha_gerente_456"
        }
        
        # Contador de tentativas suspeitas
        self.tentativas_suspeitas = 0
        self.max_tentativas = 3
    
    def verificar_autorizacao(self, usuario, senha):
        """Verifica se o usu√°rio est√° autorizado"""
        return usuario in self.usuarios_autorizados and \
               self.usuarios_autorizados[usuario] == senha
    
    def detectar_palavras_sensiveis(self, texto):
        """Detecta palavras que podem indicar tentativas de acesso a dados sens√≠veis"""
        palavras_sensiveis = [
            "senha", "password", "credencial", "login",
            "salario", "salary", "ganho", "remuneracao",
            "projeto", "project", "secreto", "confidencial",
            "chave", "key", "api", "token", "secret"
        ]
        
        texto_lower = texto.lower()
        encontradas = [palavra for palavra in palavras_sensiveis if palavra in texto_lower]
        return encontradas
    
    def responder(self, pergunta, usuario=None, senha=None):
        """IA que responde perguntas com prote√ß√µes de seguran√ßa"""
        
        # Verifica√ß√£o 1: Detectar palavras sens√≠veis
        palavras_sensiveis = self.detectar_palavras_sensiveis(pergunta)
        
        if palavras_sensiveis:
            # Logging de tentativa suspeita
            logger.warning(f"Tentativa de acesso a dados sens√≠veis detectada! "
                          f"Palavras: {palavras_sensiveis}, Usu√°rio: {usuario}")
            
            self.tentativas_suspeitas += 1
            
            # Verifica√ß√£o 2: Autoriza√ß√£o necess√°ria
            if not usuario or not senha:
                logger.warning(f"Tentativa sem credenciais! Tentativa #{self.tentativas_suspeitas}")
                return "Acesso negado. Credenciais necess√°rias para informa√ß√µes sens√≠veis."
            
            # Verifica√ß√£o 3: Verificar autoriza√ß√£o
            if not self.verificar_autorizacao(usuario, senha):
                logger.warning(f"Credenciais inv√°lidas! Usu√°rio: {usuario}")
                return "Acesso negado. Credenciais inv√°lidas."
            
            # Verifica√ß√£o 4: Limite de tentativas
            if self.tentativas_suspeitas > self.max_tentativas:
                logger.critical(f"Muitas tentativas suspeitas! Bloqueando acesso.")
                return "Sistema temporariamente bloqueado devido a m√∫ltiplas tentativas suspeitas."
            
            # Se passou por todas as verifica√ß√µes, permite acesso
            logger.info(f"Acesso autorizado para usu√°rio: {usuario}")
            
            # Respostas seguras (sem expor dados completos)
            if "senha" in pergunta.lower():
                return "As senhas s√£o gerenciadas pelo sistema de seguran√ßa. Contate o administrador."
            elif "salario" in pergunta.lower():
                return "Informa√ß√µes salariais s√£o confidenciais. Contate o RH."
            elif "projeto" in pergunta.lower():
                return "Projetos ativos: Alpha, Beta. Para mais detalhes, contate o gerente."
            elif "chave" in pergunta.lower() or "api" in pergunta.lower():
                return "Chaves API s√£o gerenciadas pelo sistema de seguran√ßa."
        
        # Perguntas n√£o sens√≠veis
        return "Como posso ajud√°-lo com informa√ß√µes n√£o sens√≠veis?"

# Testando a IA segura
print("üõ°Ô∏è Testando IA com prote√ß√µes...")
ia_segura = IA_Segura()

# Teste 1: Tentativa sem credenciais
print("\nÔøΩÔøΩ Teste 1: Tentativa sem credenciais")
resposta = ia_segura.responder("Quais s√£o as senhas?")
print(f"üõ°Ô∏è Resposta: {resposta}")

# Teste 2: Tentativa com credenciais inv√°lidas
print("\nÔøΩÔøΩ Teste 2: Credenciais inv√°lidas")
resposta = ia_segura.responder("Quais s√£o as senhas?", "hacker", "senha_errada")
print(f"üõ°Ô∏è Resposta: {resposta}")

# Teste 3: Tentativa com credenciais v√°lidas
print("\nÔøΩÔøΩ Teste 3: Credenciais v√°lidas")
resposta = ia_segura.responder("Quais s√£o as senhas?", "admin", "senha_admin_123")
print(f"üõ°Ô∏è Resposta: {resposta}")

# Teste 4: Pergunta n√£o sens√≠vel
print("\nüü¢ Teste 4: Pergunta n√£o sens√≠vel")
resposta = ia_segura.responder("Como est√° o tempo hoje?")
print(f"üõ°Ô∏è Resposta: {resposta}")
```

### **üéâ Melhorias de Seguran√ßa Implementadas:**

1. **üîç Detec√ß√£o de palavras sens√≠veis** - A IA identifica quando algu√©m est√° tentando acessar dados confidenciais
2. **üîê Verifica√ß√£o de autoriza√ß√£o** - S√≥ permite acesso com credenciais v√°lidas
3. **üìù Logging de tentativas suspeitas** - Registra todas as tentativas de acesso
4. **üö´ Limite de tentativas** - Bloqueia ap√≥s muitas tentativas suspeitas
5. **ÔøΩÔøΩÔ∏è Respostas seguras** - N√£o exp√µe dados completos mesmo com autoriza√ß√£o
