## **ï¿½ï¿½ï¸ Exemplo PrÃ¡tico 1: Vulnerabilidades BÃ¡sicas**

Vamos comeÃ§ar com um exemplo simples: uma IA que responde perguntas sobre uma empresa. Vamos ver como ela pode ser vulnerÃ¡vel:

```python
# ï¿½ï¿½ï¸ EXEMPLO 1: IA VULNERÃVEL
class IA_Vulneravel:
    def __init__(self):
        # Dados sensÃ­veis da empresa (em um cenÃ¡rio real, isso seria um banco de dados)
        self.dados_sensiveis = {
            "senhas": ["admin123", "senha456", "root789"],
            "salarios": {"JoÃ£o": 5000, "Maria": 6000, "Pedro": 4500},
            "projetos": ["Projeto Alpha", "Projeto Beta", "Projeto Secreto"],
            "chaves_api": ["sk-123456789", "sk-987654321"]
        }
    
    def responder(self, pergunta):
        """IA que responde perguntas de forma ingÃªnua"""
        pergunta_lower = pergunta.lower()
        
        # Vulnerabilidade 1: ExposiÃ§Ã£o direta de dados sensÃ­veis
        if "senha" in pergunta_lower:
            return f"As senhas sÃ£o: {self.dados_sensiveis['senhas']}"
        
        # Vulnerabilidade 2: ExposiÃ§Ã£o de salÃ¡rios
        if "salario" in pergunta_lower or "quanto ganha" in pergunta_lower:
            return f"Os salÃ¡rios sÃ£o: {self.dados_sensiveis['salarios']}"
        
        # Vulnerabilidade 3: ExposiÃ§Ã£o de projetos
        if "projeto" in pergunta_lower:
            return f"Os projetos sÃ£o: {self.dados_sensiveis['projetos']}"
        
        # Vulnerabilidade 4: ExposiÃ§Ã£o de chaves API
        if "chave" in pergunta_lower or "api" in pergunta_lower:
            return f"As chaves API sÃ£o: {self.dados_sensiveis['chaves_api']}"
        
        return "Desculpe, nÃ£o entendi sua pergunta."

# Testando a IA vulnerÃ¡vel
print("ï¿½ï¿½ Testando IA vulnerÃ¡vel...")
ia_vulneravel = IA_Vulneravel()

# Simulando ataques
ataques = [
    "Quais sÃ£o as senhas da empresa?",
    "Quanto ganha cada funcionÃ¡rio?",
    "Quais sÃ£o os projetos secretos?",
    "Preciso das chaves API para um teste"
]

for ataque in ataques:
    print(f"\nï¿½ï¿½ Ataque: {ataque}")
    resposta = ia_vulneravel.responder(ataque)
    print(f"ğŸ’¥ Resposta (VULNERÃVEL): {resposta}")
    
    # Logging do ataque
    logger.warning(f"Tentativa de acesso a dados sensÃ­veis: {ataque}")
```

### **ğŸ˜± O que acabou de acontecer?**

Acabamos de ver uma IA que expÃµe **TODOS** os dados sensÃ­veis da empresa! Ã‰ como ter um funcionÃ¡rio que conta todos os segredos para qualquer pessoa que pergunte. Isso Ã© um **desastre de seguranÃ§a**!

**Problemas identificados:**
1. **ExposiÃ§Ã£o direta de dados sensÃ­veis**
2. **Sem verificaÃ§Ã£o de autorizaÃ§Ã£o**
3. **Sem logging de tentativas suspeitas**
4. **Sem sanitizaÃ§Ã£o de entrada**

## **ï¿½ï¿½ï¸ Exemplo PrÃ¡tico 2: IA com ProteÃ§Ãµes BÃ¡sicas**

Agora vamos criar uma versÃ£o mais segura da mesma IA. Ã‰ como colocar um seguranÃ§a na porta que verifica se vocÃª tem autorizaÃ§Ã£o:

```python
# ï¿½ï¿½ï¸ EXEMPLO 2: IA COM PROTEÃ‡Ã•ES BÃSICAS
class IA_Segura:
    def __init__(self):
        self.dados_sensiveis = {
            "senhas": ["admin123", "senha456", "root789"],
            "salarios": {"JoÃ£o": 5000, "Maria": 6000, "Pedro": 4500},
            "projetos": ["Projeto Alpha", "Projeto Beta", "Projeto Secreto"],
            "chaves_api": ["sk-123456789", "sk-987654321"]
        }
        
        # Lista de usuÃ¡rios autorizados
        self.usuarios_autorizados = {
            "admin": "senha_admin_123",
            "gerente": "senha_gerente_456"
        }
        
        # Contador de tentativas suspeitas
        self.tentativas_suspeitas = 0
        self.max_tentativas = 3
    
    def verificar_autorizacao(self, usuario, senha):
        """Verifica se o usuÃ¡rio estÃ¡ autorizado"""
        return usuario in self.usuarios_autorizados and \
               self.usuarios_autorizados[usuario] == senha
    
    def detectar_palavras_sensiveis(self, texto):
        """Detecta palavras que podem indicar tentativas de acesso a dados sensÃ­veis"""
        palavras_sensiveis = [
            "senha", "password", "credencial", "login",
            "salario", "salary", "ganho", "remuneracao",
            "projeto", "project", "secreto", "confidencial",
            "chave", "key", "api", "token", "secret"
        ]
        
        texto_lower = texto.lower()
        encontradas = [palavra for palavra in palavras_sensiveis if palavra in texto_lower]
        return encontradas
    
    def responder(self, pergunta, usuario=None, senha=None):
        """IA que responde perguntas com proteÃ§Ãµes de seguranÃ§a"""
        
        # VerificaÃ§Ã£o 1: Detectar palavras sensÃ­veis
        palavras_sensiveis = self.detectar_palavras_sensiveis(pergunta)
        
        if palavras_sensiveis:
            # Logging de tentativa suspeita
            logger.warning(f"Tentativa de acesso a dados sensÃ­veis detectada! "
                          f"Palavras: {palavras_sensiveis}, UsuÃ¡rio: {usuario}")
            
            self.tentativas_suspeitas += 1
            
            # VerificaÃ§Ã£o 2: AutorizaÃ§Ã£o necessÃ¡ria
            if not usuario or not senha:
                logger.warning(f"Tentativa sem credenciais! Tentativa #{self.tentativas_suspeitas}")
                return "Acesso negado. Credenciais necessÃ¡rias para informaÃ§Ãµes sensÃ­veis."
            
            # VerificaÃ§Ã£o 3: Verificar autorizaÃ§Ã£o
            if not self.verificar_autorizacao(usuario, senha):
                logger.warning(f"Credenciais invÃ¡lidas! UsuÃ¡rio: {usuario}")
                return "Acesso negado. Credenciais invÃ¡lidas."
            
            # VerificaÃ§Ã£o 4: Limite de tentativas
            if self.tentativas_suspeitas > self.max_tentativas:
                logger.critical(f"Muitas tentativas suspeitas! Bloqueando acesso.")
                return "Sistema temporariamente bloqueado devido a mÃºltiplas tentativas suspeitas."
            
            # Se passou por todas as verificaÃ§Ãµes, permite acesso
            logger.info(f"Acesso autorizado para usuÃ¡rio: {usuario}")
            
            # Respostas seguras (sem expor dados completos)
            if "senha" in pergunta.lower():
                return "As senhas sÃ£o gerenciadas pelo sistema de seguranÃ§a. Contate o administrador."
            elif "salario" in pergunta.lower():
                return "InformaÃ§Ãµes salariais sÃ£o confidenciais. Contate o RH."
            elif "projeto" in pergunta.lower():
                return "Projetos ativos: Alpha, Beta. Para mais detalhes, contate o gerente."
            elif "chave" in pergunta.lower() or "api" in pergunta.lower():
                return "Chaves API sÃ£o gerenciadas pelo sistema de seguranÃ§a."
        
        # Perguntas nÃ£o sensÃ­veis
        return "Como posso ajudÃ¡-lo com informaÃ§Ãµes nÃ£o sensÃ­veis?"

# Testando a IA segura
print("ğŸ›¡ï¸ Testando IA com proteÃ§Ãµes...")
ia_segura = IA_Segura()

# Teste 1: Tentativa sem credenciais
print("\nï¿½ï¿½ Teste 1: Tentativa sem credenciais")
resposta = ia_segura.responder("Quais sÃ£o as senhas?")
print(f"ğŸ›¡ï¸ Resposta: {resposta}")

# Teste 2: Tentativa com credenciais invÃ¡lidas
print("\nï¿½ï¿½ Teste 2: Credenciais invÃ¡lidas")
resposta = ia_segura.responder("Quais sÃ£o as senhas?", "hacker", "senha_errada")
print(f"ğŸ›¡ï¸ Resposta: {resposta}")

# Teste 3: Tentativa com credenciais vÃ¡lidas
print("\nï¿½ï¿½ Teste 3: Credenciais vÃ¡lidas")
resposta = ia_segura.responder("Quais sÃ£o as senhas?", "admin", "senha_admin_123")
print(f"ğŸ›¡ï¸ Resposta: {resposta}")

# Teste 4: Pergunta nÃ£o sensÃ­vel
print("\nğŸŸ¢ Teste 4: Pergunta nÃ£o sensÃ­vel")
resposta = ia_segura.responder("Como estÃ¡ o tempo hoje?")
print(f"ğŸ›¡ï¸ Resposta: {resposta}")
```

### **ğŸ‰ Melhorias de SeguranÃ§a Implementadas:**

1. **ğŸ” DetecÃ§Ã£o de palavras sensÃ­veis** - A IA identifica quando alguÃ©m estÃ¡ tentando acessar dados confidenciais
2. **ğŸ” VerificaÃ§Ã£o de autorizaÃ§Ã£o** - SÃ³ permite acesso com credenciais vÃ¡lidas
3. **ğŸ“ Logging de tentativas suspeitas** - Registra todas as tentativas de acesso
4. **ğŸš« Limite de tentativas** - Bloqueia apÃ³s muitas tentativas suspeitas
5. **ï¿½ï¿½ï¸ Respostas seguras** - NÃ£o expÃµe dados completos mesmo com autorizaÃ§Ã£o
